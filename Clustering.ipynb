```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/iris-clustr/Iris.csv')
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
df1 = df[df['Species']== 'Iris-setosa']
df2 = df[df['Species']== 'Iris-versicolor']
df3 = df[df['Species']== 'Iris-virginica']
plt.scatter(df1['PetalLengthCm'],df1['PetalWidthCm'],color ='red')
plt.scatter(df2['PetalLengthCm'],df2['PetalWidthCm'],color ='black')
plt.scatter(df3['PetalLengthCm'],df3['PetalWidthCm'],color ='green')
```

```python
km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(df[['PetalLengthCm']],df[['PetalWidthCm']])
y_predicted
```

```python
km.cluster_centers_
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df['Species']== 'Iris-setosa']
df2 = df[df['Species']== 'Iris-versicolor']
df3 = df[df['Species']== 'Iris-virginica']
plt.scatter(df1['PetalLengthCm'],df1['PetalWidthCm'],color ='red')
plt.scatter(df2['PetalLengthCm'],df2['PetalWidthCm'],color ='black')
plt.scatter(df3['PetalLengthCm'],df3['PetalWidthCm'],color ='green')
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.show()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['PetalLengthCm']],df[['PetalWidthCm']])
    sse.append(km.inertia_)
    
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
